<h2> #4. Сколько у каждого из пользователей в среднем покупок в неделю (по месяцам)? Не стоит забывать, что внутри месяца может быть не целое количество недель. Например, в ноябре 2021 года 4,28 недели. И внутри метрики это нужно учесть. 

<h2/>

```python
# но у olist_customers_dataset и olist_order_items_dataset - нет общих колонок
# соединим


df_4 = pd.merge(olist_customers_dataset, pokupki_started, on = 'customer_id')
df_5 = pd.merge(df_4, olist_order_items_dataset, on = 'order_id')

#df_5= df_5.query("order_status == ['processing', 'shipped', 'delivered']")
```

```python
# для пирога нам нужно: номер уник клиента, номер уник заказа, число покупок в заказе, дата заказа, число недель в месяце заказа
# customer_unique_id, order_id, order_item_id, order_purchase_timestamp
```

```python
# найдем макс число покупок в зазазе - это будет число покупок на заказ

orders_maxs = df_5.groupby(['customer_unique_id', 'order_id','order_purchase_timestamp']).agg({'order_item_id' : 'max'}).reset_index()
orders_maxs.sort_values('order_id', ascending = False)
```

```python
# посмотрим повторяются ли покупатели, если да - это значит что у них бывает более 1-го order_id

duplicates_exist = orders_maxs['customer_unique_id'].duplicated().any()

if duplicates_exist:
    print("Дубликаты существуют в колонке 'customer_unique_id'.")
else:
    print("Дубликатов в колонке 'customer_unique_id' нет.")
```

```python
# Заказы дублируются, поэтому сгруппируем и проссумируем числа покупок на уник. клиента и по месяцу

orders_maxs_counted = orders_maxs = df_5.groupby(['customer_unique_id', 'order_purchase_timestamp']).agg({'order_item_id' : 'sum'}).reset_index()

orders_maxs_counted
```

```python
orders_maxs['order_purchase_timestamp'] = pd.to_datetime(orders_maxs['order_purchase_timestamp'])

# Группировка данных по месяцу и customer_unique_id, суммирование order_item_id
monthly_orders_sum = orders_maxs.groupby([orders_maxs['order_purchase_timestamp'].dt.to_period('M'), 'customer_unique_id'])['order_item_id'].sum().reset_index()
monthly_orders_sum
```

```python
# Создание новой колонки числом недель в месяце - с месяцем, в котором был сделан заказ

import calendar

def weeks_in_month(year, month):
    _, days_in_month = calendar.monthrange(year, month)
    return days_in_month / 7


monthly_orders_sum['weeks_in_month'] = monthly_orders_sum['order_purchase_timestamp'].apply(lambda x: weeks_in_month(x.year, x.month))

monthly_orders_sum
```

```python
# считаем сколько покупок приходится у уник. клиента в месяц

monthly_orders_sum['purchases_per_month'] = monthly_orders_sum['order_item_id'] / monthly_orders_sum['weeks_in_month']
monthly_orders_sum.sort_values('purchases_per_month')
```

<h2> #5 Когортный анализ пользователей. В период с января по декабрь выявите когорту с самым высоким retention на 3-й месяц.<h2/>

```python
df_6 = customers_orders_1
#.query("order_status == ['processing', 'shipped', 'delivered']")
df_6 = df_6.dropna()
```

```python
df_6.head()
```

```python
df_6_copy = df_6.copy()
```

```python
# Преобразование столбцов с датами в тип данных datetime
df_6_copy['order_purchase_timestamp'] = pd.to_datetime(df_6_copy['order_purchase_timestamp'])
df_6_copy['order_month'] = df_6_copy['order_purchase_timestamp'].dt.to_period('M')

# Определение месяца первой покупки для каждого клиента
df_6_copy['cohort_month'] = df_6_copy.groupby('customer_unique_id')['order_purchase_timestamp'].transform('min').dt.to_period('M')

# Расчет временного интервала с месяца первой покупки для каждой транзакции
df_6_copy['cohort_index'] = (df_6_copy['order_month'].dt.to_timestamp('M') - df_6_copy['cohort_month'].dt.to_timestamp('M')) / np.timedelta64(1, 'M')
df_6_copy['cohort_index'] = df_6_copy['cohort_index'].astype(int)
```

```python
# Группировка данных по когортам и индексам когорт
cohort_data = df_6_copy.groupby(['cohort_month', 'cohort_index']).agg({
    'customer_unique_id': 'nunique',
    'order_id': 'nunique'  # используем 'order_id' для подсчета количества заказов вместо 'payment_value'
}).reset_index()

# Переименование столбцов для удобства
cohort_data.rename(columns={
    'customer_unique_id': 'TotalCustomers',
    'order_id': 'TotalOrders',
}, inplace=True)

# Расчет Retention Rate
cohort_data['RetentionRate'] = cohort_data['TotalCustomers'] / cohort_data.groupby('cohort_month')['TotalCustomers'].transform('max')
```

```python
cohort_data_filtered = cohort_data[(cohort_data['cohort_month'] >= '2017-01') & (cohort_data['cohort_month'] <= '2017-12')]
cohort_data_filtered
```

```python
cohort_pivot =  pd.pivot_table(cohort_data_filtered, values='TotalCustomers', index='cohort_month', columns='cohort_index')
cohort_pivot
```

```python
# retention rate
cohort_pivot =  pd.pivot_table(cohort_data_filtered, values='RetentionRate', index='cohort_index', columns='cohort_month')
cohort_pivot
```

```python
# Выбор когорты с самым высоким retention на третий месяц
cohort_with_highest_retention = cohort_data_filtered[cohort_data_filtered['cohort_index'] == 2].sort_values(by='RetentionRate', ascending=False).iloc[0]

# Вывод результата
print("Когорта с самым высоким retention на третий месяц:")
print(cohort_with_highest_retention)

```

В кластеризации можешь выбрать следующие метрики:

Пример такого описания: 
    RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней,

"""

```python
# Определяем функцию для категоризации частоты покупок
def categorize_frequency(freq):
    if freq == 1:
        return 1
    elif freq == 2:
        return 2
    elif freq == 3:
        return 3
    elif freq == 4:
        return 4
    else:
        return 5

# Создаем новый столбец 'frequency_category' с категориями частоты покупок
frequency_df['frequency_score'] = frequency_df['frequency'].apply(categorize_frequency)

frequency_df.sort_values('frequency', ascending=False)
```

```python
rfm_scores = rfm_scores[['customer_unique_id',
                         'recency', 'frequency', 'monetary',
                         'recency_score', 'frequency_score', 'monetary_score']]
rfm_scores['RFM Score'] = rfm_scores['recency_score'].map(str) + rfm_scores['frequency_score'].map(str) + rfm_scores['monetary_score'].map(str)
rfm_scores
```

```python
# Функция для создания сегментов на основе оценок RFM
def create_segments(row):
    if row['recency_score'] <= 2 and row['frequency_score'] <= 2 and row['monetary_score'] >= 4:
        return 'Champions'
    elif row['recency_score'] <= 2 and row['frequency_score'] <= 2:
        return 'Loyal Customers'
    elif row['recency_score'] <= 2 and row['frequency_score'] == 5:
        return 'New Customers'
    elif row['recency_score'] <= 2 and row['frequency_score'] >= 4:
        return 'Potential Loyalists'
    elif row['recency_score'] == 3 and row['frequency_score'] <= 2:
        return 'About to Sleep'
    elif row['recency_score'] == 3 and row['frequency_score'] >= 4:
        return 'Need Attention'
    elif row['recency_score'] >= 4 and row['frequency_score'] <= 2:
        return 'At Risk'
    elif row['recency_score'] >= 4 and row['frequency_score'] == 5:
        return 'Can’t Lose Them'
    elif row['recency_score'] >= 4 and row['frequency_score'] >= 3 and row['monetary_score'] >= 3:
        return 'Promising'
    else:
        return 'Others'

# Создание столбца с сегментами на основе оценок RFM
rfm_scores['segment'] = rfm_scores.apply(create_segments, axis=1)
```

```python
# Plot the distribution of customers over R and F
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

for i, p in enumerate(['recency_score', 'frequency_score']):
    parameters = {'recency_score': 'Recency', 'frequency_score': 'Frequency'}
    y = rfm_scores[p].value_counts().sort_index()
    x = y.index
    ax = axes[i]
    bars = ax.bar(x, y, color='silver')
    ax.set_frame_on(False)
    ax.tick_params(left=False, labelleft=False, bottom=False)
    ax.set_title('Distribution of {}'.format(parameters[p]), fontsize=14)
    for bar in bars:
        value = bar.get_height()
        if value == y.max():
            bar.set_color('firebrick')
        ax.text(bar.get_x() + bar.get_width() / 2,
                value - 5,
                '{}\n({:.0%})'.format(int(value), value / y.sum()),
                ha='center',
                va='top',
                color='w')

plt.show()

# Plot the distribution of M for RF score
fig, axes = plt.subplots(nrows=5, ncols=5,
                         sharex=False, sharey=True,
                         figsize=(10, 10))

r_range = range(1, 6)
f_range = range(1, 6)
for r in r_range:
    for f in f_range:
        y = rfm_scores[(rfm_scores['recency_score'] == r) & (rfm_scores['frequency_score'] == f)]['monetary'].value_counts().sort_index()
        x = y.index
        ax = axes[r - 1, f - 1]
        bars = ax.bar(x, y, color='silver')
        if r == 5:
            if f == 3:
                ax.set_xlabel('{}\nF'.format(f), va='top')
            else:
                ax.set_xlabel('{}\n'.format(f), va='top')
        if f == 1:
            if r == 3:
                ax.set_ylabel('R\n{}'.format(r))
            else:
                ax.set_ylabel(r)
        ax.set_frame_on(False)
        ax.tick_params(left=False, labelleft=False, bottom=False)
        ax.set_xticks(x)
        ax.set_xticklabels(x, fontsize=8)

        for bar in bars:
            value = bar.get_height()
            if value == y.max():
                bar.set_color('firebrick')
            ax.text(bar.get_x() + bar.get_width() / 2,
                    value,
                    int(value),
                    ha='center',
                    va='bottom',
                    color='k')
fig.suptitle('Distribution of M for each F and R', fontsize=14)
plt.tight_layout()
plt.show()
```
